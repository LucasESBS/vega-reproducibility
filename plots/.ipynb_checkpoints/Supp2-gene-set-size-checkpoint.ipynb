```python
import sys
import os
sys.path.append('../src/')
import pandas as pd
import numpy as np
import itertools
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams['axes.linewidth'] = .5
matplotlib.rcParams['lines.linewidth'] = .5
matplotlib.rcParams['patch.linewidth'] = .5
from scipy.stats import spearmanr
import seaborn as sns
import pandas as pd
```

```python
# reload - need to run notebook for Figure 1 first to generate this file
df = pd.read_csv('../paper/supT-vae-gsea-ifn.csv', index_col=0, header=0)
```

```python
fig = plt.figure(figsize=[7,7])
plt.scatter(np.log10(df['Gene_set_size']), np.abs(df['Bayes_Factor']), s=20, linewidth=0, c='k')
plt.xlabel(r'$log_{10}$(Gene set size)', fontsize=15)
plt.ylabel(r'$|log_e|$(BF)', fontsize=15)
plt.tick_params(labelsize=15)
#plt.text(2.7, 8, s='Spearman R:%.2f'%(spearmanr(np.log10(df['Gene_set_size']), np.abs(df['Bayes_Factor']))[0]), fontsize=15)
#plt.savefig('../paper/figures/supp-gene-set-bias-vae.pdf', format='pdf', dpi=300, bbox_inches='tight')
```

```python
fig = plt.figure(figsize=[7,7])
plt.scatter(np.log10(df['Gene_set_size']), df['GSEA_-log10FDR'], s=20, linewidth=0, c='k')
plt.xlabel(r'$log_{10}$(Gene set size)', fontsize=15)
plt.ylabel(r'$-log_{10}$(FDR)', fontsize=15)
plt.tick_params(labelsize=15)
#plt.text(2.7, 4, s='Spearman R:%.2f'%(spearmanr(np.log10(df['Gene_set_size']), df['GSEA_-log10FDR'])[0]), fontsize=15)
#plt.savefig('../paper/figures/supp-gene-set-bias-gsea.pdf', format='pdf', dpi=300, bbox_inches='tight')
```
